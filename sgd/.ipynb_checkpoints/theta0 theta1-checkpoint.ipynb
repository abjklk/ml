```python
import numpy as np
import matplotlib.pyplot as plt
# Make a prediction with coefficients
import pylab
def predict(row, coefficients):
    yhat = coefficients[0]
    for i in range(len(row)-1):
        yhat += coefficients[i + 1] * row[i] 
    return yhat
```

```python
# Estimate linear regression coefficients using stochastic gradient descent

def coefficients_sgd(coeff,train, l_rate, n_epoch):
    coef = coeff
    #print(coef)
    for epoch in range(n_epoch):
        sum_error = 0.0
        for row in train:
            yhat = predict(row, coef)
            error = yhat - row[-1]
            sum_error += (error **2)         
            coef[0] = coef[0] - l_rate * error
            for i in range(len(row)-1):
                coef[i + 1] = coef[i + 1] - l_rate * error * row[i]
        print('>epoch=%d, lrate=%.4f, error=%.3f' % (epoch, l_rate, (sum_error/(2*len(train)))))
    
    cost=(sum_error/(2*len(train)))
    print("Cost: %f"%(cost))
    return cost

#what 4th line will d0
#trigger to stop
#change alpha find theta0,1
#plot epoch-error
#change theta0,1 init and plot coeff-error
```

```python
import csv

p=[]
dataset=[]
with open('6.txt') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            x = float(row[0])
            y = float(row[1])
            p =[x,y]
            dataset.append(p)
            
l_rate = 0.0001
n_epoch = 1000
theta0=[]
theta1=[]
theta=[]
costarr=[]
for i in range(0,10):
    theta0.append(i*2)
    theta1.append(i*0.1)
    theta.append([i*2,i*0.1])
for i in theta:
    sg_cost = coefficients_sgd(i,dataset, l_rate, n_epoch)
    costarr.append(sg_cost)
    
print(costarr)
print(theta)
```
