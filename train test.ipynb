```python
import numpy as np
import matplotlib.pyplot as plt
# Make a prediction with coefficients
import pylab
def predict(row, coefficients):
    yhat = coefficients[0]
    for i in range(len(row)-1):
        yhat += coefficients[i + 1] * row[i] 
    return yhat
```

```python
# Estimate linear regression coefficients using stochastic gradient descent

def coefficients_sgd(train, l_rate, n_epoch):
    coef = [0.0 for i in range(len(train[0]))]
    #print(coef)
    for epoch in range(n_epoch):
        sum_error = 0.0
        for row in train:
            yhat = predict(row, coef)
            error = yhat - row[-1]
            sum_error += (error **2)         
            coef[0] = coef[0] - l_rate * error
            for i in range(len(row)-1):
                coef[i + 1] = coef[i + 1] - l_rate * error * row[i]
        print('>epoch=%d, lrate=%.4f, error=%.3f' % (epoch, l_rate, (sum_error/(2*len(train)))))
        err=(sum_error/(2*len(train)))
    return coef,err

#what 4th line will d0
#trigger to stop
#change alpha find theta0,1
#plot epoch-error
#change theta0,1 init and plot coeff-error
```

```python
import csv
from sklearn.model_selection import train_test_split
p=[]
dataset=[]
x=[]
y=[]

trainset=[]
testset=[]
with open('6.txt') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            a = float(row[0])
            x.append(a)
            b = float(row[1])
            y.append(b)
            p =[a,b]
            dataset.append(p)
            
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
l_rate = 0.0001
n_epoch = 1000
for i in range(len(X_train)):
    g=[X_train[i],y_train[i]]
    trainset.append(g)
for i in range(len(X_test)):
    d=[X_test[i],y_test[i]]
    testset.append(d)
coef,err = coefficients_sgd(trainset, l_rate, n_epoch)
coef1,err1 = coefficients_sgd(testset, l_rate, n_epoch)
print("Coef of train",coef)
print("Coef of test",coef1)
print("Cost of Trainset is",err)
print("Cost of Testset is",err1)
```

```python
samples=len(dataset)
m=[]
n=[]
y_predict=[]
for i in range(samples):
    m.append(dataset[i][0])
    n.append(dataset[i][1])
    q=coef[0] + coef[1]*m[i]
    y_predict.append(q)
    

pylab.plot(m,n,'o')
pylab.plot(m,y_predict,'k-')
pylab.show()
print(coef)
```
