```python
import numpy as np
import matplotlib.pyplot as plt
# Make a prediction with coefficients
import pylab
def predict(row, coefficients):
    yhat = coefficients[0]
    for i in range(len(row)-1):
        yhat += coefficients[i + 1] * row[i] 
    return yhat
```

```python
# Estimate linear regression coefficients using stochastic gradient descent

def coefficients_sgd(train, l_rate, n_epoch):
    coef = [0.0 for i in range(len(train[0]))]
    #print(coef)
    for epoch in range(n_epoch):
        sum_error = 0.0
        for row in train:
            yhat = predict(row, coef)
            error = yhat - row[-1]
            sum_error += (error **2)         
            coef[0] = coef[0] - l_rate * error
            for i in range(len(row)-1):
                coef[i + 1] = coef[i + 1] - l_rate * error * row[i]
        print('>epoch=%d, lrate=%.4f, error=%.3f' % (epoch, l_rate, (sum_error/(2*len(train)))))
    
    cost=(sum_error/(2*len(train)))
    print("Cost: %f"%(cost))
    return cost,coef

#what 4th line will d0
#trigger to stop
#change alpha find theta0,1
#plot epoch-error
#change theta0,1 init and plot coeff-error
```

```python
import csv

p=[]
dataset=[]
with open('6.txt') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            x = float(row[0])
            y = float(row[1])
            p =[x,y]
            dataset.append(p)
            
l_rate = [0.00005,0.00008,0.0001,0.0005,0.0008,0.001,0.005,0.008,0.01,0.05]
n_epoch = 1000

costarr=[]
for i in l_rate:
    sg_cost,coef = coefficients_sgd(dataset, i, n_epoch)
    costarr.append(sg_cost)

# print(costarr)
print(l_rate)
print(costarr)
```

```python
plt.plot(l_rate,costarr)
plt.xlabel('learning rate')
plt.ylabel('J ()')
plt.title("LR vs Cost")
plt.show()
```
